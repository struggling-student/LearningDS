```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
paziente1 = pd.read_csv('./paziente1.csv', sep=';')
paziente2 = pd.read_csv('./paziente2.csv', sep=';')
```

```python
# columns to drop since they have 0 values, useless for the analysis.
# NOTE: I could also delete the .state columns since they don't represent any movement but I'll keep them for now)
columns_to_drop=[
    'FootLeft.rx','FootLeft.ry','FootLeft.rz',
    'ThumbLeft.rx','ThumbLeft.ry','ThumbLeft.rz',
    'HandTipLeft.rx','HandTipLeft.ry','HandTipLeft.rz',
    'ThumbRight.rx','ThumbRight.ry','ThumbRight.rz',
    'HandTipRight.rx','HandTipRight.ry','HandTipRight.rz',
    'FootRight.rx', 'FootRight.ry', 'FootRight.rz',
    'Head.rx','Head.ry','Head.rz',
    # also remove timestamp and datetime
    'timestamp', 'datetime'
]
# clean data
def clean_data(df):
    clean_df = df.drop(columns_to_drop, axis=1)
    return clean_df
clean_paziente1 = clean_data(paziente1)
clean_paziente2 = clean_data(paziente2)

clean_paziente1
```

```python
def get_data(df):
    data = df[[
    'SpineMid.px', 'SpineMid.py', 'SpineMid.pz',
    ]]
    return data

spine_paziente1 = get_data(clean_paziente1)
spine_paziente2 = get_data(clean_paziente2)

def plot(pz1, pz2):
    ax = plt.axes(projection='3d')
    ax.plot3D(pz2['SpineMid.px'], pz2['SpineMid.py'], pz2['SpineMid.pz'], color='green')
    ax.plot3D(pz1['SpineMid.px'], pz1['SpineMid.py'], pz1['SpineMid.pz'], color='red')
    plt.show()
plot(spine_paziente1, spine_paziente2)
```
